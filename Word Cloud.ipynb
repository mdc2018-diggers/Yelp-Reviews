```python
def top_words(dataset_name='review', limit=10000, max_words=100, stop_words=stop_words):
    wordcount = defaultdict(int)
    for text in dataset.read(dataset_name, attrs='text', limit=limit):
        for word in re.findall(r'(\w+)', text):
            word = word.lower()
            if word not in stop_words:
                wordcount[word] += 1

    top_words = {
        word: count
        for count, word in sorted([(count, word) for word, count in wordcount.items()], reverse=True)[:max_words]
    }
    return top_words
top_words('review', max_words=10)
```

```python
from highcharts import Highchart

chart = Highchart()
chart.set_options('title', {'text': 'Word Cloud'})
data = [
    {'name': word, 'weight': count}
    for word, count in top_words(max_words=100).items()
]
chart.add_data_set(data, series_type='wordcloud', name='Example Series')
chart
```

```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt

wordcloud = WordCloud(max_font_size=50).generate_from_frequencies(top_words())
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from PIL import Image
import numpy as np

image = np.array(Image.open('digger.png'))

wordcloud = WordCloud(max_font_size=50, mask=image).generate_from_frequencies(top_words('tip'))
wordcloud_colors = ImageColorGenerator(image)
plt.figure()
plt.imshow(wordcloud.recolor(color_func=wordcloud_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```
